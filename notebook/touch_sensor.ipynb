```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("../results/touch_sensor.csv")
df.head()
```

```python
plt.figure(figsize=(10, 5), dpi=144)
ax=plt.subplot(111)
#ax.set_xlim(235, 310)
ax.plot(list(range(0, len(df.left_force))), df.left_force, label = "Left Foot")
ax.plot(list(range(0, len(df.right_force))), df.right_force, label = "Right Foot")
#ax.bar(list(range(0, len(df.sum_force))), df.sum_force, label = "SUM Force")

plt.legend()
plt.show()
```
