```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("../results/simple_runner.csv")
df.head()
```

```python
plt.figure(figsize=(10, 5), dpi=144)
plt.plot(list(range(0, len(df.ground_sensor))), df.ground_sensor, label = "Ground Sensor")
plt.plot(list(range(0, len(df.central_sensor))), df.central_sensor, label = "Central Sensor")
plt.plot(list(range(0, len(df.outer_left_sensor))), df.outer_left_sensor, label = "Outer Left Sensor")
plt.legend()
plt.show()
```
