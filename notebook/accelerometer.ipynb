```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("../results/accelerometer.csv")
df.head()
```

```python
plt.figure(figsize=(10, 5), dpi=144)
ax=plt.subplot(111)
#ax.set_xlim(235, 310)
ax.plot(list(range(0, len(df.x_axis))), df.x_axis, label = "X Axis")
ax.plot(list(range(0, len(df.y_axis))), df.y_axis, label = "Y Axis")
ax.plot(list(range(0, len(df.z_axis))), df.z_axis, label = "Z Axis")

ax.annotate("Breaking", xy=(53, -3), xytext=(55, 25), arrowprops={"arrowstyle":"->", "color":"gray"})
ax.annotate("Acceleration", xy=(153, 3), xytext=(155, 25), arrowprops={"arrowstyle":"->", "color":"gray"})

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5), dpi=144)
ax=plt.subplot(111)
ax.set_xlim(235, 310)
ax.plot(list(range(0, len(df.x_axis))), df.x_axis, label = "X Axis")
ax.plot(list(range(0, len(df.y_axis))), df.y_axis, label = "Y Axis")
ax.plot(list(range(0, len(df.z_axis))), df.z_axis, label = "Z Axis")

plt.plot([261, 261], [-100, 100], 'k-', lw=1)
plt.plot([265, 265], [-100, 100], 'k-', lw=1)
plt.plot([280, 280], [-100, 100], 'k-', lw=1)
plt.plot([284, 284], [-100, 100], 'k-', lw=1)

ax.annotate("Falling dowm", xy=(260, 10), xytext=(250, 30), arrowprops={"arrowstyle":"->", "color":"gray"})
ax.annotate("Impact on floor", xy=(280, 10), xytext=(268, 30), arrowprops={"arrowstyle":"->", "color":"gray"})

plt.legend()
plt.show()
```
