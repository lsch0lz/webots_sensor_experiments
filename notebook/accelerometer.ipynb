```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("../results/accelerometer.csv")
df.head()
```

```python
plt.figure(figsize=(10, 5), dpi=144)
ax=plt.subplot(111)
#ax.set_xlim(235, 310)
ax.plot(list(range(0, len(df.x_axis))), df.x_axis, label = "X Axis")
ax.plot(list(range(0, len(df.y_axis))), df.y_axis, label = "Y Axis")
ax.plot(list(range(0, len(df.z_axis))), df.z_axis, label = "Z Axis")

ax.annotate("Breaking", xy=(102, -3), xytext=(110, -25), arrowprops={"arrowstyle":"->", "color":"gray"})
ax.annotate("Acceleration", xy=(153, 3), xytext=(155, 25), arrowprops={"arrowstyle":"->", "color":"gray"})

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5), dpi=144)
ax=plt.subplot(111)
ax.set_xlim(245, 330)
ax.plot(list(range(0, len(df.x_axis))), df.x_axis, label = "X Axis")
ax.plot(list(range(0, len(df.y_axis))), df.y_axis, label = "Y Axis")
ax.plot(list(range(0, len(df.z_axis))), df.z_axis, label = "Z Axis")

ax.annotate("Falling down", xy=(255, 15), xytext=(260, 30), arrowprops={"arrowstyle":"->", "color":"gray"})
ax.annotate("1st Impact on floor", xy=(275, 30), xytext=(260, 50), arrowprops={"arrowstyle":"->", "color":"gray"})
ax.annotate("2nd Impact on floor", xy=(295, 30), xytext=(280, 50), arrowprops={"arrowstyle":"->", "color":"gray"})
ax.annotate("3rd Impact \n on floor", xy=(304, 33), xytext=(298, 50), arrowprops={"arrowstyle":"->", "color":"gray"})
ax.annotate("Driving again", xy=(325, 15), xytext=(315, 50), arrowprops={"arrowstyle":"->", "color":"gray"})



plt.legend()
plt.show()
```
