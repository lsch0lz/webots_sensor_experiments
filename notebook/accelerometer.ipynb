```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("../results/accelerometer.csv")
df.head()
```

```python
plt.figure(figsize=(10, 5), dpi=144)
plt.plot(list(range(0, len(df.x_axis))), df.x_axis, label = "X Axis")
plt.plot(list(range(0, len(df.y_axis))), df.y_axis, label = "Y Axis")
plt.plot(list(range(0, len(df.z_axis))), df.z_axis, label = "Z Axis")
plt.legend()
plt.show()
```
